```python
df = pd.read_csv('temp_dw_ai_normal_deliver_range_aoi_basic_chengdu_20200929_1601451326255.txt')
```

```python
def encode(row):
    pair = gis.gcj02_to_wgs84(row['lng'], row['lat'])
    return geohash.encode(pair[1],pair[0], precision=6)

df['geohash'] = df.apply(lambda row: encode(row), axis=1)
```

```python
for index, value in df[df['district_name'].isin(['青羊区', '金牛区', '武侯区', '成华区'])].iterrows():
    print(value['id'], value['name'])
    
    if value['id'] in ['B001C8MJL3']:
        continue
    
    partial = pd.read_csv('spider.csv')
    
    try:
        if len(partial) > 0 and value['id'] in partial['id'].values:
            continue

        result = get_store_list(value['lng'], value['lat'], uuid, access_token)
        print(result)

        if 'store_list' not in result.keys() or result['store_list'] is None:
            print('error')
            break
        
        m = df[df['geohash'] == value['geohash']][['id', 'geohash']]

        if len(result['store_list']) > 0:

            if result['store_list'][0]['distance'] <= 2000:
                m['is_dispatch'] = 1
                partial.append(m, ignore_index=True).to_csv('spider.csv', index=False)
            else:
                m['is_dispatch'] = 0
                partial.append(m, ignore_index=True).to_csv('spider.csv', index=False)
        else:
            m['is_dispatch'] = 0
            partial.append(m, ignore_index=True).to_csv('spider.csv', index=False)
    except:
        pass

    time.sleep(75)
```
